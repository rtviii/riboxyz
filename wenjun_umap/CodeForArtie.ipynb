```python
import numpy as np

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy as sp
# from sklearn.manifold import MDS
import sys


%load_ext autoreload
%autoreload 2

plt.rcParams.update({'font.size': 20})  # Sets global font size

import umap
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


K_GW_full = np.load('Tunnel_GW_normalize.npy')
K_GW_full = K_GW_full/K_GW_full.max()

# Fit UMAP with 3 components
fit = umap.UMAP(metric="precomputed", n_components=2, random_state=52)
mapper = fit.fit_transform(K_GW_full)

# Plot the 3D projection
fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111)#, projection='3d')

ax.scatter(mapper[:, 0], mapper[:, 1],  s=20)

ax.set_xlabel("UMAP 1")
ax.set_ylabel("UMAP 2")

plt.tight_layout()
# plt.savefig('GW_shapes.png')
```
